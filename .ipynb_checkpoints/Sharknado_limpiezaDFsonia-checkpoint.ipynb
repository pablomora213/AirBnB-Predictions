```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sharknado = pd.read_csv('data/attacks.csv',encoding="ISO-8859–1")
sharknado
```

```python
sharknado.isnull().sum()
```

```python
#Miramos y seleccionamos que variables(columnas) queremos tener
sharknados = sharknado[["Case Number", "Date", "Year","Type", "Country", "Area", "Activity", "Sex ", "Age", "Injury", "Fatal (Y/N)", "Species "]]
sharknados
```

```python
#Modificamos los nombres de las variables para que podamos trabajar con facilidad
df = sharknados.copy()
df.columns
```

```python
renombre= [a.replace(" ", "_").lower() for a in df.columns]
renombre
```

```python
df.columns= renombre
df
```

```python
df.rename(columns={'case_number': 'case', 'sex_': 'sex', "fatal_(y/n)": "dead", "species_": "especie" }, inplace=True)
df
```

```python
str(df['age'][3]).lower()
```

```python
def resolucion(x):
    x=str(x).lower()
    if "no injury" in x:
        x= "no injury"
        return x
    elif "fatal" in x:
        x= "death"
        return x
    elif x == "nan":
        return np.nan
    else:
        x= "bite"
        return x
```

```python
df["resolucion"]= df.apply(resolucion)
df
```
