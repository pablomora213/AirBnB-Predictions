```python
# Funciones auxiliares

patron='[+-]?([0-9]*[.])?[0-9]+|[0-9]+'

def limpieza_ba(x):
    ''' pasar a float el numero de baños de una stirng'''
    n_ba=0
    
    try:
        n_ba=float(re.match(patron, x)[0])
    except:
        n_ba=0
        
    return n_ba
#propertyclean=dfcolumns.property_type.value_counts()[:16]
#propertyclean=list(propertyclean.keys())
#p=propertyclean
p= ['Entire rental unit',
 'Private room in rental unit',
 'Entire residential home',
 'Entire condominium (condo)',
 'Private room in bed and breakfast',
 'Room in boutique hotel',
 'Entire loft',
 'Entire townhouse',
 'Private room in residential home',
 'Entire serviced apartment',
 'Private room in houseboat',
 'Houseboat',
 'Room in hotel',
 'Private room in guest suite',
 'Private room in townhouse',
 'Boat']

def propclean (x):
    ''' para categorizar la columna y poder hacer dummys después'''
    x=str(x)
    if x in p and x!='Nan':
        return x
    else:
        return 'OTHER'

def outliersprice (x):
    ''' me quita los outliers segun el miniestudio hecho de la columna'''
    if x> 1000:
        return 500
    elif x== 0:
        return 167.8
    else:
        return x
```

```python
def limpieza_dataset(data):
    '''todas las funciones de limppieza agrupadas para el train, como lleva price no puede pasar por el test'''
    
    
    df=data[['id','price', 'host_acceptance_rate','host_is_superhost','host_listings_count', 
             'neighbourhood_cleansed', 'latitude', 'longitude', 'property_type','room_type', 'accommodates', 'bathrooms_text',
             'bedrooms', 'beds', 'minimum_nights', 'number_of_reviews_ltm','review_scores_rating', 'review_scores_accuracy', 
             'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
             'review_scores_value', 'reviews_per_month']]
    
    dfcolumns=df.drop(columns=['beds', 'bedrooms', 'reviews_per_month', 'review_scores_accuracy', 
                               'review_scores_communication', 'review_scores_rating'])
    
    dfcolumns['bathroom']=dfcolumns.bathrooms_text.apply(limpieza_ba) #extraer numeros del string
    
    dfcolumns['host_acceptance_rate']=dfcolumns.host_acceptance_rate.apply(limpieza_ba) #sacar los numeros del porcentaje, q es sting
    #dfcolumns.host_acceptance_rate.replace(0, np.nan, inplace=True) no mejora cambiando los ceros por la media
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.room_type).reset_index(drop=True)], axis=1) #dummys
    
    dfcolumns.property_type = dfcolumns.property_type.apply(propclean) #meter palabras dentro del diccionario o pasarlas a other
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.property_type).reset_index(drop=True)], axis=1)
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.neighbourhood_cleansed).reset_index(drop=True)], axis=1)
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.host_is_superhost).reset_index(drop=True)], axis=1)
    
    dfcolumns['price']=dfcolumns.price.apply(outliersprice) #quitar outliers de price
 
    
    dfcolumns=dfcolumns.drop(columns=['bathrooms_text', 'room_type', 'property_type', 'neighbourhood_cleansed', 'host_is_superhost'])
    #quitar las columnas que hemos generado dummys
    
    ##trabajamos los nan
    #dfcolumns=dfcolumns.dropna()  #no en test!!!!
    #dfcolumns=dfcolumns.fillna(0)  # funciona mejor la media que ceros por nan
    
    dfcolumns=dfcolumns.apply(lambda x: x.fillna(x.mean()),axis=0)

    return dfcolumns
```

```python
def limpieza_dataset2(data):
    '''mismo que el anterior pero sin la columna price'''
    
    
    df=data[['id', 'host_acceptance_rate','host_is_superhost','host_listings_count', 
             'neighbourhood_cleansed', 'latitude', 'longitude', 'property_type','room_type', 'accommodates', 'bathrooms_text',
             'bedrooms', 'beds', 'minimum_nights', 'number_of_reviews_ltm','review_scores_rating', 'review_scores_accuracy', 
             'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
             'review_scores_value', 'reviews_per_month']]
    
    dfcolumns=df.drop(columns=['beds', 'bedrooms', 'reviews_per_month', 'review_scores_accuracy', 
                               'review_scores_communication', 'review_scores_rating'])
    
    dfcolumns['bathroom']=dfcolumns.bathrooms_text.apply(limpieza_ba) #extraer numeros del string
    
    dfcolumns['host_acceptance_rate']=dfcolumns.host_acceptance_rate.apply(limpieza_ba) #sacar los numeros del porcentaje, q es sting
    #dfcolumns.host_acceptance_rate.replace(0, np.nan, inplace=True) no mejora cambiando los ceros por la media
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.room_type).reset_index(drop=True)], axis=1) #dummys
    
    dfcolumns.property_type = dfcolumns.property_type.apply(propclean) #meter palabras dentro del diccionario o pasarlas a other
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.property_type).reset_index(drop=True)], axis=1)
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.neighbourhood_cleansed).reset_index(drop=True)], axis=1)
    
    dfcolumns=pd.concat([dfcolumns, pd.get_dummies(dfcolumns.host_is_superhost).reset_index(drop=True)], axis=1)
    
    #dfcolumns['price']=dfcolumns.price.apply(outliersprice) #quitar outliers de price
 
    
    dfcolumns=dfcolumns.drop(columns=['bathrooms_text', 'room_type', 'property_type', 'neighbourhood_cleansed', 'host_is_superhost'])
    #quitar las columnas que hemos generado dummys
    
    ##trabajamos los nan
    #dfcolumns=dfcolumns.dropna()  #no en test!!!!
    #dfcolumns=dfcolumns.fillna(0)  # funciona mejor la media que ceros por nan
    
    dfcolumns=dfcolumns.apply(lambda x: x.fillna(x.mean()),axis=0)

    return dfcolumns
```

# Automatizar Modelos

```python
X_train_d, X_test_d, y_train_d, y_test_d = tts(dfcolumns.drop('price', axis=1), 
                                               dfcolumns.price)
```

```python
def regre(modelo):
    
    modelo.fit(X_train_d, y_train_d)
    
    train_score=modelo.score(X_train_d, y_train_d)  # R2
    test_score=modelo.score(X_test_d, y_test_d)
    
    print(modelo)
    print('Train R2:', train_score)
    print('Test R2:', test_score)
    
    return modelo
```

```python
from sklearn.linear_model import SGDRegressor as SGDR
sgdr=SGDR(max_iter=50000)

sgdr=regre(sgdr)
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNNR
knnr=KNNR(n_neighbors=5, weights='distance')

knnr=regre(knnr)
```

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR
gbr=GBR(n_estimators=50, learning_rate=0.2)

gbr=regre(gbr)  #Cogeremos este modelo porque tiene el test más alto, y menos difeencia entre el train y el test, aunque el R2 del train sea muy bajo
```

```python

from catboost import CatBoostRegressor as CTR
ctr=CTR(verbose=0)

ctr=regre(ctr)
```
